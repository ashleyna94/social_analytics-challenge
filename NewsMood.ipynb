```python
# Dependencies
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
from config import (consumer_key, consumer_secret, access_token, access_token_secret)

# Initialize sentiment analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

#Setup Tweepy API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
test = api.search("@BBC", count=1, result_type="recent")
test
```

```python
# Specify the news stations
target_terms = ["@BBCNews", "@CBSNews", "@CNN", "@FoxNews", "@nytimes"]

# Make a list to hold each type of sentiment
news_sentiment = {
    "tweet_user_list": [],
    "tweet_text_list":[], 
    "tweet_date_list": [],
    "compound_list": [], 
    "positive_list": [], 
    "negative_list": [], 
    "neutral_list": []
}

# Loop through each news station to get the recent 100 tweets
for x in range(1):

    # Loop through all news stations
    for target in target_terms:

        public_tweets = api.search(target, count=5, result_type="recent")

        # Loop through all tweets
        for tweet in public_tweets["statuses"]:

            # Grab the username, text, and date of each tweet 
            tweet_user = tweet["user"]["screen_name"]
            tweet_text = tweet["text"]
            tweet_date = tweet["created_at"]
            
            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]

            # Save all info to the corresponding lists
            news_sentiment["tweet_user_list"].append(tweet_user)
            news_sentiment["tweet_text_list"].append(tweet_text)
            news_sentiment["tweet_date_list"].append(tweet_date)
            news_sentiment["compound_list"].append(compound)
            news_sentiment["positive_list"].append(pos)
            news_sentiment["neutral_list"].append(neu)
            news_sentiment["negative_list"].append(neg)

# Create a DataFrame to hold all the info
sentiment_summary = pd.DataFrame(news_sentiment)
```
